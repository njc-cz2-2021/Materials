## 16.1.1 Local Area Network (LAN)
**Local Area Network** (LAN) is a network connecting computers on a single site, could be single room or single building.

The benefits of computer connected by LAN:
- the expense of installing application software on each individual PC could be saved by installing the software on an application server attached to the LAN instead
- a file server could be attached to the LAN that allowed users to store larger files and also allowed files to be shared between users
- Instead of suppling individual printers to be connected to a user's PC, one or more printers could be attached to a print server that was connected to the LAN; these could be higher quality printers.
- Digitization of documents and emails to be stored in file servers made possible

## 16.1.2 Wide Area Network (WAN)

The benefits of computer connected by WAN:
- a 'job' could be run on a remote computer that had the required application software 
- a data archive that was stored on a remote computer can be accessed
- a message could be transmitted electronically to a user on a remote computer

In a network, the specification on how the various devices in the network are connected are called a **topology**. We categorize topologies based on the *physicality* of the network connection and the subtypes is categorized based on the shapes of the connection.

Based on their *physicality*, there are two types of topology to consider:
- **physical network topology** refers to the *physical actual architecture of the network*, i.e how the different components in the network connected to one another physically
    The subtypes are:
    - **physical bus topology** where each end-system (termed **client**) are connected to a single cable called **backbone**
    <br>

    |----------|-------------|
    |Inexpensive to install as minimum length of cable is required| Packets are sent through the shared backbone, allowing every client on the network to see packets that aren’t intended for them.|
    |As compared to the physical star topology (see below), there is no central hub, reducing the chances of a network failure and decreasing the cost of installation.| The backbone is used for communication by multiple clients, introducing the risk of collisions.|
    ||The entire network becomes unusable in the event that the backbone is damaged.|

    </center>
    <br>
        
    - **physical star topology** where each clients in the network is *linked to a central device* (called **hub**). The cental deviced (called **hub**) receives packets for all of the clients connected to it and is responsible for delivering them to the correct recipient.
    <br>
    <center>

    |Advantages|Disadvantages|
    |----------|-------------|
    |Packets are sent directly to their recipient, over a cable that is connected only to the recipient. Other clients on the network cannot see packets that aren’t intended for them.| The failure of central hub will cause all communication over the network to be stopped|
    |It is easy to add and remove clients to and from the network.|Expensive to install due to the amount of cable required.|
    |Each cable has just one device communicating over it, eliminating the possibility of collisions.||
    |The failure of one cable does not affect the performance of the rest of the network.|| 


    </center>

- **logical network topology** refers to the *flow of data packets within a network*. 
    - in **logical bus topology**, data packets are delivered to all clients on the network
    - in **logical star topology**, data packets are delivered packets only to their recipient.

Note that physical and logical topological can be mixed. Even though a network is set up as a physical star, it can still behave as a logical bus. Similarly, on the other hand, even if the physical connections between clients and the central hub follow that of the physical star topology, running a bus protocol on the hub allows it to distribute packets to all of the connected clients so as to act like a bus network. 

# 16.3 Type of Communication between Computers in a Network

The type of communication between computers in a network is categorized into 2 types:
- **Client-server Networking** is an architecture where a client requests services provided by dedicated machines (which are usually more powerful than client machines), called a **servers**, on a network. Client machines could     sometimes act as servers. Servers are connected to the network in the same way as client.
    The clients on the network request services from the servers, which then respond to the client with the requested service. 
    
    *Examples* of circumstances where client-server approach is used:
    - the server stores a database which is accessed from the client system.
    - the server stores a web application which allows the client system to find or, sometimes, supply information
    - the server stores a web application which allows the client system to carry out an e-commerce or financial transaction
    
    *Benefits*
    - Different servers can provide for wide range of services
    - Provides for central storage for all files and documents
    - Security is centralised

    *Drawbacks*
    - initial set-up costs for the server and client is high
    - server-based network requires the use of a specialist network operating system
    - if the server fails then the work of all users is affected.
    <br>

- **Peer-to-peer network (P2P)** is a network where each peer (networked device) can act as a client to request a file from another peer or it can act as a server when another peer requests the download of a file.

    *Examples*
    - Bittorrent file-sharing 

    *Benefits*
    - no dedicated server is required
    - additional computers can join the P2P network easily
    - Each computer can act as both a provider and receiver of resources
    - more cost effective than client-server as there is no need for a powerful server to provide services

    *Drawbacks* 
    - Clients which provide servies must be running in order for the network to be fully operational
    - Decentralized system making it hard to administer
    - less control of identification of malware from other computers in the network.

# 15.4 Methods of Communication in a Network
There are two methods of communication in a network, they are **packet switching** and **circuit switching**.
- **Packet switching** is a method of data transmission where data is grouped into smaller chunks called **packets** before being transmitted over a digital network. You can think of a packet as an envelope in the postal system, they’re labelled with addresses for their sender and recipient and contain information intended for the recipient. In a network using packet switching, one message is frequently split into multiple packets, each of which is sent to its recipient via the best possible route before being reassembled with other packets by its recipient. 

    When a packet is sent through a network, it usually has to pass through a number of hardware devices called **routers** before reaching its destination. A router uses the recipient address on a packet to determine where to send the packet. Every time that a packet passes through a router, we call it a **hop**. 

    Each packet can only pass through a finite number of hops. A packet’s **time to live** (or TTL) is a number that indicates how many hops the packet can partake in and is reduced by one with each hop. When a packet’s TTL expires, the packet is said to be **dropped**, meaning that the packet is **deleted**. The recipient will notice a missing packet and request that the sender transmits the missing packet again. 

    Main components of a packet

    <center>

    |Component of packet| Purpose|
    |-------------------|--------|
    |Sender’s address| Identifies where the packet was sent from, and therefore where the response should be sent to.|
    |Receiver's address| Identifies the packet’s intended recipient, allowing it to be routed to the correct device.|
    |Packet contents| Where the packet holds the data that is being transferred. |
    |Time to live (TTL)| Holds the number of hops a packet can go through before being dropped.|
    |Sequence number | Contains the number of packets in a message and identifies a packet’s position in relation to others. This allows packets to be reassembled in the correct order and allows missing packets to be identified.

    </center>

    The internet uses packet-switching to transmit data.

    *Benefits*
    - Dedicated channel does not need to be setup before transmission takes place
    - Different routes are possible for data transfer

    *Drawbacks*
    - The biggest limitation of packet switching is that it is unsuitable for applications that require minimal latency. 
    - If there is too much traffic then packets will be dropped in transit

- **Circuit switching** is a method of implementing a telecommunications network in which two network nodes establish a dedicated communications channel (circuit) through the network before the nodes may communicate. Exmaple of circuit-switched network is the early analog telephone network. When a call is made from one telephone to another, switches within the telephone exchanges create a continuous wire circuit between the two telephones, for as long as the call lasts.

# 16.5 The Internet
The **internet** is global network of interconnected computer networks which uses end-to-end **communication protocol**(see what this means below) and *open to public*.

In contrast, **intranet** is a computer network for sharing information, collaboration tools, operational systems, and other computing services *within an organization*, usually *private*, i.e. exclude access by outsiders. 

Personal computers or devices are connected to LANs, which in turn connects to WANs, which then connects to the internet/intranet through services provided by Internet Service Providers (ISP). In Singapore, examples of ISPs are Singtel, Starhub, M1, MyRepublic and ViewQwest.

There are several physical devices that works together to connect an end-device (e.g a PC) to the Internet.
- A **switch** is a hardware that connects multiple networked devices in the network. It maintains an address table containing an unique identifier for each device connected to it, called **Media Access Control (MAC)** addresses. Switches manage the flow of data across a network by transmitting a received network packet only to the one or more devices for which the packet is intended. In other words, a switch connects multiple devices to create a network.
    >  MAC addresses is actually provided by **network interface controller (NIC)** which is a piece of hardware in your network devices responsible for converting digital data from the sending device into these signals or translating back into the corresponding data at the receiving end. MAC addresses are 48 bits long and usually referred to in hexadecimal, e.g., `52-3E-AA-DF-E4-A5`. Windows user can type `ipconfig /all` in command prompt to see the MAC addresses. 
- A **wireless access point** is a networking hardware device that allows other WiFi enabled devices to connect to a wired network. 
- A **router** is a network device that forwards packets from one network to another. Based on internal **routing tables**, which stores IP addresses and gets frequently updated, routers read each incoming packet and decide how to forward it in the most efficient/fastest possible route.  This might be the route that involves the lowest number of hops or the route that is least congested at the time. 
- A **modem** (modulator-demodulator) is a computer hardware device that converts data from a digital format into a format suitable for an analog such as telephone or radio. It connects your LAN network to your ISP. The modem takes signals from your ISP and translates them into signals your local devices can use, and vice versa. Each modem has an assigned public IP address that identifies it on the Internet.

> Before switches, a device called a **hub** was used. However, instead of directing traffic to intended devices, hubs would simply repeat messages to *all* connected devices. This slows down the network as data are sent unnecessarily to unintended recipients.

Beyond the devices mentioned earlier, there are a few more hardware that are sometimes needed in a entwork.
- A **gateway** is a piece of networking hardware or software used in telecommunications for telecommunications networks that allows data to flow from one discrete network to another, where the networks are running different, incompatible protocol. Gateways function is to translate one protocol to another. E.g one network uses 64 bit and the other 128 bit, or connecting Ethernet and WiFi.

    Where two networks use different protocols, packets must be modified so as to conform to both protocols. This is where gateways come in, they strip away most of the packet’s details, leaving just the packet’s contents. The gateway will then give packets new sender and receiver addresses which comply to the new protocol. 

- A **bridge** is hardware that connect two network segments. It direct packets from a node in a network segment to another node on a *different* segment. Unlike router, bridge stores MAC addresses connected to it. Think about it as two port switch. 

## 16.5.1 Communication Protocol
A communication **protocol** is a set of rules for data transmission which are agreed by sender and receiver so that the data is correctly transmitted. It allows devices to identify and make connections with each other, as well as formatting rules that specify how data is packaged into messages sent and received. Some other things that protocol defines:
- positive voltage represents a bit with value 0
- transmission speed that a sender must not exceed

The absence of this protocol could cause unwanted behaviours. For example, without a protocol, a transmitting computer, for example, could be sending its data in 16-bit segments while the receiving computer might expect the data in 8-bit segments, and thus, receiving computer will read the information wrongly.

TLDR, communication protocols are needed to:
- establish rules to enable communication to take place
- initiate and terminate communication

The protocol used by the Internet is Transmission Control Protocol / Internet Protocol (TCP/IP). TCP/IP will be further elaborated in later section.

## 16.5.2 Data Communication in the Internet 
### 16.5.2.1 Internet Protocol Addresses
As mentioned earlier, internet is a huge interconnected computer networks and as such, to get data from an end-point to another, we need to have a way to identify these end-points. This is where the IP address come into play. 

To define the source and destination of data in a transmission in a computer network, **IP (Internet Protocol)** addressing is used. Currently, the Internet uses IP version 4 (IPv4) addressing (Although IPv6 also exist and is used concurrently at the moment.) It uses 32 bit (4 bytes) where a group of bits is used to defined a network (**netID**) and another group is used to define a host on the network (**hostID**). Their aim was to assign a unique, universally recognised address for each device on the internet. Each of the computers in a network shares the same network identifier but has its own host identifier.

> The IP Address `127.0.0.1` is special. It is used to access the network services that are running on the device itself, i.e. any device will use `127.0.0.1` to identify itself.

In the scheme, different entities are categorized into different classes (A to E). Different classes allows for different number of unique hostIDs.

<center>

| Class | Class Identifier (first bits in the first byte) | Number of bits for netID | Number of bits for hostID |
|-------|-------------------------------------------------|--------------------------|---------------------------|
| A     | 0                                               | 7                        | 24                        |
| B     | 10                                              | 14                       | 16                        |
| C     | 110                                             | 21                       | 8                         |

</center>

For example, `10101100 00010000 00000100 00101111` is an IPv4 address. Since the first 2 bits in the first octet is `10`, it means that the host resides in a class B network. 

To make it simpler for the user, IPv4 address are actually written in denary where the denary values for each byte is separated by dots, e.g. the IPv4 address above is displayed as `172.16.4.47`.

IPv6 addresses are formed of eight blocks separated by colons. Each block contains four hexadecimal characters (`a-z` and `0-9`), e.g., `2071:0eb8:85a3:8a2f:0000:0000:0370:7264`. IPv6 addresses use 128 bits which, compared to IPv4’s use of 32 bits.

The allocation of IP addresses is managed by **internet registry**. There are only five internet registries in operation, each serving a different geographical area. They are:
1. The African Network Information Center (AFRINIC) serves Africa.
2. The American Registry for Internet Numbers (ARIN) serves Antarctica, Canada, parts of the Caribbean, and the United States.
3. The Asia-Pacific Network Information Centre (APNIC) serves East Asia, Oceania, South Asia, and Southeast Asia.
4. The Latin America and Caribbean Network Information Centre (LACNIC) serves most of the Caribbean and all of Latin America.
5. The Réseaux IP Européens Network Coordination Centre (RIPE NCC) serves Europe, Central Asia, Russia, and West Asia

An important part of an internet registry’s work is to protect the world’s depleting pool of unallocated IP addresses. When a new IP address is requested, instead of assigning a brand new IP right away, it will first check if a previously allocated IP address that has become unused can be assigned instead. 

In practice, however, if every device that is connected to a network had its own IP address, there wouldn’t be enough IP addresses to go around. As such, we make further distinction : **public** and **private** IP addresses. A **public** IP address identifies you to the wider internet so that all the information you're searching for can find you. This task is carried out by the internet registries. On the other hand, a **private** IP address is used within a private network to connect securely to other devices within that same network. *Each device within the same network has a unique private IP address.* This task is carried out by the router in your network. 

Furthermore, as the number of available private IP addresses within a private network is limited, assigning each device on a network its own private IP address would not be sensible, as that device may leave and never join again, resulting in a wasted IP address. 

As such, a protocol called **Dynamic Host Configuration Protocol (DHCP)** is used to assign private IP addresses to devices as they join a network. DHCP uses a pool of available private IP addresses to allocate these IP addresses to new devices for the duration of their session. Once a device leaves the network, the IP address that the device was using is returned to the pool for allocation to a new device. Nowaadays, most routers have the ability to function as DHCP server.

We get into a snag here, how then can a computer in a private network communicate with another computer through the internet? This is where another magic of routers plays a part. When a device on the private network needs to communicate with a device on the Internet, it sends packets *through the router*, which makes a record of the packet before replacing the private IP address of the computer with its own public IP address. When a response is received, it is sent to the router’s public IP address, which then forwards the response to the correct private IP address by using the record it made when sending the packet. 

This method of mapping an IP address space into another by modifying network address information in the IP header of packets while they are in transit across a traffic routing device is called **Network Address Translation**.

**Subnetting** is the practice of dividing a network into two or more smaller networks, called **subnets**. It allows for a more efficient use of a hostID by applying a structure to it. For example, subnetting a medium-sized organisation's network of 150 employees. 

On the internet, all of the allocated IP addresses have netID pointing to the router. The router then has to interpret the hostID to direct the transmission to the appropriate hosts on one of the networks via a gateway.

The network identifier part of an IP address can be determined with a **subnet mask**.

Let’s say that a device has an IP address of 192.168.3.24 and a subnet mask of 255.255.255.0. 
- First convert both the IP address and subnet mask into their binary form
- For every corresponding bit in each of them, do an `and` operation
- The resulting bit is the bit value of the network identifier at that position 
- For ease of reading, convert back to denary values

`IP address of device (in binary) : 11000000.10101000.00000011.00011000 `<br>
`Subnet mask          (in binary) : 11111111.11111111.11111111.00000000 `<br>
`Network identifier   (in binary) : 11000000.10101000.00000011.00000000  `<br>

The more bits that are assigned to the network identifier of an IP address, the more different subnets a network can have. In the same way, the more bits that are assigned to the host identifier, the more different devices can be connected to each subnet simultaneously. 

### 16.5.2.2 Domain Name System (DNS)

If a domain name server does not have a record of the domain that you are trying to access, your request will be passed to another domain name server. ### 16.5.3 Transmission Control Protocol / Internet Protocol (TCP / IP) Model
The protocol is used in all parts of the Internet to enable different devices to communicate. It comprises of 4 layers that sit, one below the other, in a *protocol stack*. 

> If you search around, sometimes this model is described to have 7 layers instead. They're different. The 7 layers model is the Open Systems Interconnection model (OSI model) which is a general model that characterises and standardises the communication functions of a telecommunication or computing system without regard to its underlying internal structure and technology. 

#### 16.5.3.1 Layer 4. Application Layer
The functions of the application layer is to selects and uses the correct protocol to transmit data. The layer interacts with the user with application software like a web browser, email clients etc.

These applications are identified with an integer called **port numbers**. Port numbers can run from 0 to 65353. However, ports 0 to 1023 are reserved for some common applications. 

This layer uses a number of protocols, the main among which are as follows:
- **Hyper Text Transfer Protocol(HTTP)** − It is the underlying protocol for world wide web. It defines how hypermedia messages are formatted and transmitted. Port number 80.
- **Hyper Text Transfer Protocol(HTTPS)** - Performs the same function as HTTP, but encrypts information during transmission. Port number 443
- **File Transfer Protocol(FTP)** − It is a client-server based protocol for transfer of files between client and server over the network. In particular, it handles the case when the end-systems have different operating systems with different file systems. Port number 20 & 21.
- **Simple Mail Transfer Protocol(SMTP)** − It lays down the rules and semantics for sending emails. As more emails are now web-based, SMTP has largely been replaced by HTTP to access the email application. Port number 25, 587
& 465.
- **Post Office Protocal version 3(POP3)** and **Internet Message Access Protocol(IMAP)** are both protocols to receive emails. The distinction is the location for which the emails reside, for POP3, the emails are downloaded to the client computer whereas for IMAP, the emails aren't downloaded but kept on the server while being accessible from the client.  Port number 110 & 995 for POP3, 143 for IMAP.
- **Secure Shell (SSH)** Used for remote management of computers, allowing computers to be accessed and controlled remotely. In order to access a remote computer, SSH requires a username / password combination and encrypts information during transmission. Port number

#### 16.5.3.2 Layer 3. Transport Layer 
The functions of the transport layer are:
- establishes an end to end connection between the sender and the receiver, termed **virtual path**. 
- splits the transmission into packets. Each packet has a sequence number which identifies a packet’s position in relation to other packets that form part of the same message. Also contained in each packet is the port number to be used which identifies the protocol in use.

This layer uses a number of protocols, the main among which are as follows:
- **Transmission Control Protocol (TCP)** − It is a reliable connection-oriented protocol that transmits data from the source to the destination machine without any error. A connection is established between the peer entities prior to transmission. 
    - At the sending host, TCP divides an incoming byte stream into segments and assigns a separate sequence number to each segment. TCP also manages flow control so that a fast sender does not overwhelm a slow receiver. 
    - At the receiving host, TCP reorders the segments and sends an acknowledgment to the sender for correct receipt of segments. 
    - TCP has to ensure that any response is directed back to the application protocol. 
    E.g., World of Warcraft and Final Fantasy XIV are both MMORPGs that uses TCP mainly.
- **User Datagram Protocol (UDP)** − It is a message-oriented protocol that provides a simple unreliable, connectionless, unacknowledged service. It is suitable for applications that do not require TCP’s sequencing, error control or flow control. It is used for transmitting a small amount of data where the speed of delivery is more important than the accuracy of delivery. For example, real-time connections like video streaming, VoIP, and FPS games like Overwatch and CS:GO uses UDP.

#### 16.5.3.3 Layer 2. Internet (Network) Layer
The function of the internet layer is to ensure correct routing over the internet. It provides the correct IP addresses for each packet’s source and destination. Routers work within the network layer, using the IP addresses on a packet to send it to its destination. TLDR: Source and destination IP addresses are added at this stage.

The main protocol used in this layer is Internet Protocol(IP). The protocol takes the packet received from the transport layer and adds a further header. The header contains IP addresses of both send and the receiver. To find the IP address,the DNS service to find the address corresponding to the URL supplied in the user data is likely to be used.

The packet created by IP is termed **datagram** which is then to sent to the next layer, **Data Link** Layer. 

#### 16.5.3.4 Layer 1. Link Layer
The function of link layer is to:
- Controls physical connections between pieces of hardware in a network.
- Adds MAC addresses to packets which it receives from the network layer. 

When you're sending data over the internet, your data will pass through all the layers from 4 to 1, in descending order. At every layer 3 to 1, additional information to the data that is necessary for that layer to carry out its task is added. This method of modular communication is called **encapsulation**. It can be likened to taking the data from the layer above and placing it in an envelope, writing some additional information on the envelope, then passing the envelope to the layer below. 

On the other hand, at the receiving end, your data will still pass through all the layers but now in reverse order. The reverse process of stripping away the additional information is termed **decapsulation**

# 16.6 Computer Security Threats

## 16.6.1 Types of Malwares and Effective Defences against Them
Malicious software (malware) is created with the intent to:
- cause damage to computer systems by deleting or corrupting files,
- steal data, or
- causing denial of service (see below)

As such computer users must take steps to mitigate these threats. There are many different forms of malware, including a wide variety of 
- **viruses**, malicious programs that requires a host program to replicate by writing its own code into the host program. When the host program is run, the written virus program is executed first, carrying out its intended malicious effect. It can spread when someone transfers an infected file from one computer to another, or by the virus inserting a copy of itself into another program. E.g ILOVEYOU virus 2000
- **worms**, programs that can *replicate itself* in order to spread to other computers *without the need of host programs*. Usually, the worm takes advantage of shared storage, your email service, or networking features to spread unaided around the network, e.g. Mydoom worm in 2004
- **trojans**, programs that look like legitimate program while hiding its malicious intent. Many modern trojans work by giving unathorized access of your devices to the creator of the trojan. Trojans generally don't try to replicate themselves, e.g. Zeus trojan 2007
- **spyware**, programs with malicious aims to gather information about a person or organization and send it to another entity in a way that harms the user. A particular type of spyware known as a **keylogger** can keep a record of all keys pressed, meaning that usernames and passwords are recorded and transmitted back to the malicious software's author. E.g. GO keyboard 2017

One of aims of the malwares is to cause a **denial of service (DOS) attacks** on servers. In such DOS attacks, devices that are infected by the malwares form a collection called **botnet**. The attacker behind the botnet can then coordinate the computers in the botnet to bombard a server with requests that seemed legitimate. As such, the server could easily get overwhelmed and any real request to the server cannot be processed, and as such the request for the service get denied. 

Many of the malwares exploit vulnerabilities in the computer system. As such, to guard effectively against them, we should :
- improve the **code quality** within applications used, e.g. making sure *buffer overflow* don't occur 
- test new codes thoroughly and getting another programmer to do **code review**,
- keep up to date with vulnerability patches.

##  16.6.2 Network Security Measures
Computer networks allow many users to share information and we have seen from the previous section, that things shared are not necessarily benign. To make sure that the networks are secure, the following measures can be implemented:

- **Usernames and passwords**:  These combination have been used for many years as a method of authentication on computer systems. **Authentication** is the process of proving that a user is who they say they are. 
    Password chosen should be complex so that it doesn't get easily guessed via a brute force attack, where passwords are repeatedly entered until the correct one is found.
    Password used shouldn't be reused for obvious reasons. 
    
- **Firewall** : piece of software or hardware that monitors traffic in and out of a network or stand-alone computer. Firewall can allow or block the data entering or leaving the network or device via a set of rules termed **packet filters** defined by network admistrator. There are two such approach 
    - **static filtering**: filtering is done based on the packet header fields, which include source IP address, destination IP address, port number, and protocol. However, once the connection is made, the **state** of the connection are not monitored. As such, this is also termed **stateless inspection**
    - **dynamic filtering**: incoming and outgoing traffic is continuously monitored even after connection is established. Besides the header, packet types are also analysed and the content are inspected for suspicious activity. As such, this filtering is also called **stateful inspection**

- **Intrusion Detection System (IDS)**:  a device or software application that monitors a traffic for malicious activity or policy violations and sends alert on detection. It checks real time traffic for traffic patterns or signatures of attack and generates alerts to the network administrators. 

- **Intrusion Prevention System (IPS)**: kind of similar to IDS, IPS is a device that inspects and classifies traffic. The main difference is IPS proactively tries to stop malicious traffic, instead of just alerting the network adminsitrators. It works by inspecting real time traffic and checking for traffic patterns or signatures of attack and then prevents the attacks when it is detected.

The measures outlined above aren't silver bullets. They also have their own limitations.
- password: if legit login credentials are used by malicious attackers, the attacks can still go through
- firewall: encrypted traffic carrying malicious codes can still get through
- IDS/IPS might raise many false alarms and need frequeent updates to be be effective against common attack techniques.

## 16.6.3 Encryption
Another layer on security in making sure data communicated through the internet is safe from malicious attackers is through encryption.

**Encryption** is the process of converting a message into another message via some techniques so that even if the message is intercepted, it cannot be understood readily. The original message is called **plaintext** and the converted message is called **ciphertext**. The opposite process of retrieving the original message is called **decryption**

</center>

To convert a plaintext into a ciphertext, we use an algorithm which converts Plaintext into Ciphertext. Such algorithm used is called the **cipher**. Examples of encryption algorithms include 
- the enigma machine
- caesar cipher
- the Rivest cipher 6 (RC6) cipher, and 
- the industry standard, *Advanced Encryption Standard (AES)*. 

We can think of a cipher as a function $C$ that takes parameters *plaintext* $p$ and **key** $k$. The key $k$ is a part of the formula inside the cipher that contributes to the obscuring of plaintext and recovery of the ciphertext. 

As such, the key is used at 2 points, the sender and the receiver. Thus, for the receiver to be able to decrypt the message, the sender must somehow communicate the key to the recipient. This **key exchange** is a potential weak point in the process; if an attacker intercepts the key during the key exchange and the encryption method known, the message can easily be decrypted.

### 16.6.3.1 Symmetric Encryption
In this scheme of encryption, the key used at both the sender and receiver are the same. The major weakness with this happens when the key falls into the wrong hands, the interceptor would be able to decrypt the message.

**Diffie-Hellman method** is an example to generate symmetric keys between the sender and receiver.The exact implementation is not required for A-Level. However, it is important to note that in this method, the symmetric keys are generated based on shared public information. 

### 16.6.3.2 Asymmetric Encryption
As the name suggest, in this encryption scheme, the key that is used to encrypt a message cannot be used to decrypt the ciphertext. This overcomes the weakness of the symmetric encryption mentioned previously. 

One such example is the **RSA algorithm** (named after Rivest, Shamir, and Adleman, who devised it) generates several components. These components are mathematically linked so that a message encrypted using one set of components can be decrypted using the other set. The components are known collectively as a **public and private key pair**. As such asymetric encription is called termed **public-key encryption**.

In this scheme, if sender and receiver wants to communicate securely, each of need their own public/private key pair, i.e. sender has the key pair $(k^{\text{private}}_{\text{sender}},k^{\text{public}}_{\text{sender}})$ while the receiver has $(k^{\text{private}}_{\text{receiver}},k^{\text{public}}_{\text{receiver}})$. The public keys can be shared openly; they are of no use to anyone else. However, each person must keep their own private key absolutely secret.

When *sender sends a message*, the key used to encrypt is the *receiver's public key* $k^{\text{public}}_{\text{receiver}}$. When the message *arrived at the receiver*, the decryption uses the *receiver's private key* $k^{\text{private}}_{\text{receiver}}$.

So far, this process allows the sender to be sure that only the intended receiver can read the sent message. However, since public key is public, everyone knows receivers's public key, so the receiver can't be sure who the sender was.

However, this problem can be solved using the same scheme. The trick is to do another round of encryption before the ciphertext get sent to the receiver. In the other words:
1. Sender encrypts the message to receiver using the *sender's own private key* $k^{\text{private}}_{\text{sender}}$,
2. The ciphertext in Step 1 is further encrypted with *receiver's public key* $k^{\text{public}}_{\text{receiver}}$
3. When the message *arrived at the receiver*, *receiver's private key* $k^{\text{private}}_{\text{receiver}}$ is used to decrypt to reveal the ciphertext produced in Step 1.
4. The receiver then decrypts the ciphertext with *the sender's public key* $k^{\text{public}}_{\text{sender}}$, revealing the original plaintext.

## 15.6.4 Digital Signatures and Digital Certificate
A **digital signature** is a way to authenticate sender of message and guarantee the integrity of a message. Previously, we have seen how to authenticate the sender of the message. To guarantee the integrity, we add another step before step 1 in the previous section and modify some of the steps.

0. The plaintext is hashed with a hash function to produce what is called a **hash total** or **digest**.
1. The digest is then encrypted with *sender's own private key* $k^{\text{private}}_{\text{sender}}$. This encrypted digest is the **digital signature** of the message. 
2. The digital signature is then added to the message. 
3. Sender encrypts the message and the digital signature using the *sender's own private key* $k^{\text{private}}_{\text{sender}}$,
4. The ciphertext in Step 3 is further encrypted with *receiver's public key* $k^{\text{public}}_{\text{receiver}}$
5. When the message *arrived at the receiver*, *receiver's private key* $k^{\text{private}}_{\text{receiver}}$ is used to decrypt to reveal the ciphertext produced in Step 3.
6. The receiver then decrypts the ciphertext with *the sender's public key* $k^{\text{public}}_{\text{sender}}$, revealing the original plaintext and the digital signature.
7. The digital signature is then decrypted with *sender's public key* $k^{\text{public}}_{\text{sender}}$
8. The receiver runs the hash function on the plaintext and compare it with the decrypted digest. If they match, the receiver was not changed in transit and can be sure of the message's integrity.. This last process is termed verification of the digital signature.

However, anyone can create a digital signature and claim that they are a trusted individual. 

A **digital certificate** is another layer of authentication for the identities of a message sender or a website. It is issued by an official **certificate authority (CA)**, such as Let's Encrypt or Verisign. A digital certificate includes a serial number, expiry date, the holder's name, and the holder's public key. It will be 'signed' by the digital certificate of the issuing CA, so that you can verify that it is genuine.

# 16.7 Socket Programming with Python
Previously, in TCP/IP model section on the Application layer, we have learnt about IP addresses and port numbers. 

A **socket address** is just a combination of an IP address and a port number, e.g., `110.32.24.7:80`. 

Development of application programs that makes use of these socket address to communicate is called **socket/network programming**.

The most common type for socket applications are client-server applications, where we have a server part and a client part. As the name suggests:
- the server part wait for requests from the clients
- clients request services from the servers.

## 16.7.1 Server Program
To make a server program in Python, we carry out the following in sequence:
1. import `socket` module, 
2. create a `socket` object,
3. assign an IP address and port number to the socket object using the `bind()` method. `bind()` takes in a tuple of IP address and port number `(ip,port)`. If `ip=''`, then the socket will listen from *any computer* in the network. Take note that port numbers 0 to 1023 should be avoided as they're reserved. In A-Level, your're only required to create a server program that communicates with interacts with a client program on the same computer, so `ip= '127.0.0.1'`, which refers to localhost. 
4. put the server into listening mode using `listen()` method, i.e. this allows the server to be active and wait for incoming connections. `listen()` accepts an integer `n` as parameter, where `n` is the number of connections that are kept waiting when the server is busy. If there's another incoming connection during that busy time, it will be refused.
5. using a while loop, we accept incoming connections to that socket using `accept()` method. `accept()` returns a tuple of `connection` object and ip address.
6. with the `connection` object from part 5, we can now either:
    - receive message using `recv()` method. The parameter is an integer indicating maximum amount of data to be received at once. To be safe, it is recommeded by the official documentation to use `4096`. The return value is a `bytes` object, an immutable sequence of integers in the range $0 <= x < 256$, representing the data received. As such, it is important that is received data is converted back with `decode()` method.  
    - send message using `send()` method. `send()` takes in `bytes` object as parameters, as such we need to encode the data first using `encode()` method.
7. process the data as needed
8. close the connection by using `close()` method on the connection object.

Below in an example of a Python server program that does nothing but send the data sent to it back to the client. Such servers are called **echo server**.

```python
# STEP 1
from socket import socket			

# STEP 2
s = socket()		

# STEP 3
port = 12345			
s.bind(('', port))		

# STEP 4
s.listen(5)	

# STEP 5
while True:

    # STEP 6. c is the connection object
    c, addr = s.accept()	

    data = c.recv(4096)
    data = data.decode()
    
    # STEP 7
    # This is where your processing happens
    processed_data = data

    # STEP 6
    c.send(processed_data.encode())

    # STEP 8
    c.close()

    # Breaking once connection closed
    break
```

# 16.8 Verification Check on Transferred Data 
When data is transmitted, a verication check on the data when it is received is performed called the **checksum check**. Its purpose is to detect errors that may have been introduced during its transmission.

In general, checksum check works as follows:
- locate the byte(s) designed as the **checksum byte(s)** in the transmitted data packet
- calculate the sum of the number of values of the remaining bytes in the data packet. This sum is called the **checksum value**
- compare **checksum value** (in binary) with the checksum byte(s). 
- if the checksum value matches with the checksum byte, we conclude that that the data packet has been successfully transmitted error. 

## Example
Consider the following packet of size 8 bytes. The last byte (byte 7) transmitted store the checksum.

<center>

| Byte 7        | Byte 6       | Byte 5        | Byte 4        | Byte 3        | Byte 2        | Byte 1        | Byte 0        |
|----------|---------|----------|----------|----------|----------|----------|----------|
| `10110100` | `1000000` | `00010000` | `00000011` | `00001100` | `00000000` | `00010001` | `00000100` |

</center>

Checksum value: $128+16+3+12+0+17+4=180=(10110100)_{2}$, which is equal to the checksum byte.

> Qn: What is a weakness of doing checksum check?

Assuming the data is transferrred in bytes using a seven-bit code, another alternative to verify the data received is **one-bit parity check**. For this check, there are two different parities that can be implemented:
- **odd parity**, means that the sum of bits in each transmitted bytes is an odd number, i.e., the number 1s in a byte must be odd
- similarly, **even parity** just means that the sum of bits in each transmitted bytes is an even number, i.e., the number of 1s in a byte is even.

Assuming we are using *even parity* check, we have the following procedure.
1. At the transmitting end, the number of 1s in the seven-bit code is counted.
2. If the count gives an even number, the last bit in the byte, termed the **parity bit** is set to 0. Else, the parity bit is set to 1.
3. Repeat this for every byte in the transmission,
4. At the receiving end, the number of 1s in the eight-bit code is counted.
5. If the count gives an even number, the byte is accepted,
6. Repeat for every byte in the transmission. 

Note that for odd parity check, step 2 and 5 in the procedure above is adjusted accordingly.

Therefore, when used, one-bit parity check can tell us which byte contains incorrect bits.

> Qn: What is a weakness of doing one-bit parity check?

## Parity Block Check
To overcome the weakness mentioned above, **parity block check** approach can be used. In this check, before transmission
1. outgoing bytes are arranged in rectangles, where each row in the rectangle represents the bits in the byte transmitted,
2. for each column in the rectangle, we will perform the one-bit parity check. The parity bits are then bundled inside a byte termed **parity byte**,
3. transmit the data bytes followed by the parity byte. 

## Example
Consider the following example. We have four data bytes, followed by a parity byte. The parity check used here is odd parity.

<center>

|      | Bit 7 | Bit 6  | Bit 5  | Bit 4  | Bit 3  | Bit 2  | Bit 1  | Bit 0 |
|-------------|---|---|---|---|---|---|---|---|
| Byte 1      | 1 | 1 | 0 | 0 | 1 | 1 | 0 | 1 |
| Byte 2      | 0 | 0 | 1 | 0 | 0 | 0 | 0 | 0 |
| Byte 3      | 1 | 0 | 1 | 1 | 1 | 1 | 0 | 0 |
| Byte 4      | 1 | 1 | 0 | 0 | 0 | 0 | 1 | 0 |
| Parity byte | 0 | 1 | 1 | 0 | 1 | 1 | 0 | 0 |
